```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
D=pd.read_csv("../test-data", sep=";", header=None)

grouped = {}
for row in D.iterrows():
    key = row[1][0]
    value = float(row[1][1])
    if not key in grouped:
        grouped[key] = []
    grouped[key].append(value)
```

```python
names = list(grouped.keys())
box_plot_data = list(grouped.values())

fig, ax = plt.subplots(figsize=(6, 2))

#fig.title("Fruit Growth Distribution")
ax.boxplot(box_plot_data, labels=names, whis=(0.5,99.5), showfliers=False, vert=0, widths=0.8, sym='.', flierprops=dict(markersize=4))
# ax.set_xlim([-75,-64])
# ax.set_xticks(list(range(-90, -60, 1)))
plt.grid(visible=True, which='major', axis='x', color='grey', linestyle='--', linewidth=0.5)
#ax.set_yticks()
plt.plot();
```
